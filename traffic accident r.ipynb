แสดงจำนวนอุบัติเหตุแยกตามประเภทความเสียหาย <br>
จากกราฟจำนวนผู้เสียชีวิตแสดงให้เห็นว่า ช่วงปลายปีถึงต้นปี(ธันวาคม - มกราคม) มีจำนวนผู้เสียชีวิตมากกว่าช่วงเวลาอื่น <br>
ในขณะที่กราฟจำนวนผู้ได้รับบาตเจ็บและพิการไม่มีลักษณะที่แตกต่างกันในระหว่างปี 

```r
options(repr.plot.width=7)
options(repr.plot.height=4)
p <- ggplot(df, aes(x = month, y = dead, group = year, shape=factor(year), color=factor(year)))
p <- p + geom_line() + geom_point() + scale_x_discrete() + theme(legend.position='top')
p
```

จากกราฟจำนวนผู้เสียชีวิตเป็นที่น่าแปลกใจว่า เมื่อเวลาผ่านไป จำนวนผู้เสียชีวิตกลับเพิ่มสูงขึ้น สังเกตจากเส้นกราฟปี 2010 อยู่ต่ำกว่าปีอื่นๆ

```r
options(repr.plot.width=7)
options(repr.plot.height=4)
p <- ggplot(data = df, aes(x = month, y = injured, group = year, shape=factor(year), color=factor(year)))
p <- p + geom_point() + geom_line() + scale_x_discrete() + theme(legend.position='top')
p <- p + geom_line(data=aggregate)
p
```

เช่นเดียวกันกับจำนวนผู้บาดเจ็บ จำนวนผู้บาดเจ็บในปี 2010 ต่ำกว่าปีอื่นๆทั้งหมด

```r
options(repr.plot.width=7)
options(repr.plot.height=4)
p <- ggplot(data = df, aes(x=month, y=disabled, group=year, shape=factor(year), color=factor(year)))
p <- p + geom_point() + geom_line() + scale_x_discrete() + theme(legend.position='top') 
p
```

แต่จำนวนผู้พิการ มีแนวโน้มลดลง

```r
for(i in 1:12){
    if (i != 4){
        tt <- t.test(dead~month, df[df$month==4 | df$month==i,])
        print(tt$p.value)
    }
}
```

```r
for(i in 1:12){
    if (i != 12){
        tt <- t.test(dead~month, df[df$month==12 | df$month==i,])
        print(tt$p.value)
    }
}
```

```r
for(i in 1:11){
    if (i != 11){
        tt <- t.test(dead~month, df[df$month==11 | df$month==i,])
        print(tt$p.value)
    }
}
```
