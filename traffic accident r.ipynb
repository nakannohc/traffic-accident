```r
library(ggplot2)
library(repr)
```

```r
df <- read.csv('https://docs.google.com/feeds/download/spreadsheets/Export?key=1ck74o6FUXPWSH2YW4Q-VMjvvXQ_hKu9RHugAKfOBYM4&exportFormat=csv&gid=0', 1)
```

แสดงจำนวนอุบัติเหตุแยกตามประเภทความเสียหาย <br>
จากกราฟจำนวนผู้เสียชีวิตแสดงให้เห็นว่า ช่วงปลายปีถึงต้นปี(ธันวาคม - มกราคม) มีจำนวนผู้เสียชีวิตมากกว่าช่วงเวลาอื่น <br>
ในขณะที่กราฟจำนวนผู้ได้รับบาตเจ็บและพิการไม่มีลักษณะที่แตกต่างกันในระหว่างปี 

```r
options(repr.plot.width=7)
options(repr.plot.height=4)
p <- ggplot(data = df, aes(x = month, y = dead, group = month, shape=factor(year), color=factor(year)))
p <- p + geom_point() + scale_x_discrete() + theme(legend.position='top')
p
```

```r
options(repr.plot.width=7)
options(repr.plot.height=4)
p <- ggplot(data = df, aes(x = month, y = injured, group = month, shape=factor(year), color=factor(year)))
p <- p + geom_point() + scale_x_discrete() + theme(legend.position='top')
p
```
